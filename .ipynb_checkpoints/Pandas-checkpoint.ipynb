## Data Frames

```python
df = pd.DataFrame({
    "Name":["Furkan","Giri","Preet","Diya","Evan"],
    "dept": ["MLAI","Finance","MLAI","IT","HR"],
    "Salary":[52000,48000,47999,75000,50000],
    "City": ["Bharuch","Vapi","Surat","Delhi","Mumbai"]
})
```

```python
print(df)
```

```python
print(df.shape)
print(list(df.columns))
df.info()
df.describe()
df.head()
```

```python
raw = pd.DataFrame({
    "order_id":["1001","1002","1003"],
    "amount": ["2500","1800","3200"], 
})
```

```python
print(raw["amount"].sum())
```

```python
clean = raw.copy()
```

```python
clean["order_id"] = pd.to_numeric(clean["order_id"])
clean["amount"] = pd.to_numeric(clean["amount"])

print(clean["amount"].sum())
```

```python
messy = pd.DataFrame({
     "product" : ["pen","book","bag","lamp"],
     "price" : [20,np.nan,850,1200],
     "rating": [4.1,np.nan,4.5,4.5]
 })
print(messy)
```

```python
messy.isna()
```

```python
messy.isna().sum()
```

```python
filled = messy.copy()
```

```python
filled["price"] = filled["price"].fillna(filled["price"].mean())
```

```python
filled["rating"] = filled["rating"].fillna(filled["rating"].median())
```

```python
print(filled)
```
