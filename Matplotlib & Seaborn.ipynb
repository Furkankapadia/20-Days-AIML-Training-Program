## Matplotlib

figure : Canva

Axes : individual plots inside the figure

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y, label = "sin(x)", color = "green")

ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_title("Sine_Wave")
ax.legend()
plt.show()
```

```python
np.random.seed(42)
epochs = np.arange (1,21)
train_loss = np.exp(-epochs/5) + np.random.rand(20)*0.5
```

```python
val_loss = np.exp(-epochs/6) + np.random.rand(20)*0.08
```

```python
plt.plot(epochs, train_loss, label = "train")
plt.plot(epochs, val_loss, label = "Validity")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

## Bar Plot

Use for the categorical data

```python
models = ["LogReg","RF","SVM","XGB"]
acc = [0.82,0.88,0.85,0.91]
```

```python
plt.bar(models,acc,color = "grey")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model comparision")
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins = 30, color ="yellow", edgecolor = "black")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
```

```python
## Subplot

fig, axes = plt.subplots(1,2, figsize = (10,4))
axes[0].plot(epochs,train_loss)
axes[1].hist(data, bins=30)
```

```python
## BOX PLOT

sns.boxplot(data = iris, x = "species", y = "petal_length")
plt.title("Box Plot")
plt.show()
```
